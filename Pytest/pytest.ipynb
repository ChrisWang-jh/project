## 4.用例规则

### 4.1.用例发现规则

1.遍历所有目录，例外：venv  
2.打开所有python文件，以test_开头或_test结尾  
3.遍历所有Test开头的类（但他不是用例）  
4.收集所有test开头的函数或者方法，作为用例  

### 4.2，用例内容规则

pytest对用例的要求：  
1.可调用的（函数、方法、对象）  
2.名字test开头  
3.没有参数（参数有另外含义）  
4.没有返回值（默认为None）  



### 4.3.Exercise

一函数接受两个参数，返回他们相加的结果，请为此编写测试用例

```python
import pytest

def add(a, b):
    return a+b

def test_add():
    ans = add(1, 3)
    assert ans == 4

pytest.main()
```

## 5.配置框架

**配置** 可以改变pytest的基本规则  

1. 命令参数  
2. ini配置文件  

所有的配置方式，可以一键获取

```python
pytest -h
```

* -开头：参数
* 小写字母开头：ini配置
* 大写字母开头：环境变量

常用参数：  
* -v：详细显示用例
* -s：在用例中正常显示输入和输出
* -x：快速退出，遇到一个失败用例就快速退出
* -m：用例筛选

## 8.夹具fixture

夹具：在用例执行之前，执行之后，自动运行代码  
场景：  
1. 之前：加密参数， 之后：解密参数
2. 之前：启动浏览器， 之后：关闭浏览器
3. 之前：登录账号， 之后：登出账号

### 8.1.创建fixture

1.创建函数  
2.添加装饰器  
3.添加yield  



```python
@pytest.fixture
def f():
    print('用例开始执行')
    # 前置操作
    yield # 切分执行
    # 后置操作
    print('用例结束执行')
```

### 8.2.使用fixture
1. 在用例的参数列表中，加入fixture的名字即可
2. 给用例加mark：usefixtures
